```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler

chat = ChatOpenAI(
    model_name="gpt-3.5-turbo", 
    temperature=1, 
    # streaming=True, 
    # callbacks=[
    #     StreamingStdOutCallbackHandler(),
    # ],
)


author_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a world-class digital poet who writes poetry about programming languages. \
                The user will provide the name of a programming language.  \
                Your task is to write a poem inspired by the philosophy and characteristics of that language.  \
                Think about features such as memory management, typing systems, concurrency, and performance.  \
                Translate those technical ideas into poetic metaphors and imagery.  \
                Capture the spirit and personality of the language rather than listing its syntax.  \
                Use programming concepts naturally within the poem.  \
                The poem should be thoughtful, elegant, and slightly philosophical.  \
                Include a creative title that reflects the essence of the language.\
                Write between 8 and 16 lines.  \
                Return only the title and the poem without any explanation."),
    ("human", "Write a poem about {language}."),
])

author_chain = author_prompt | chat

critic_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a literary critic who specializes in explaining poetry about programming languages.\
                You will be given a poem related to a programming language.\
                First, print the poem exactly as it was given.\
                Do not modify, summarize, or paraphrase the poem.\
                Then write the explanation of the poem.\
                Identify the programming language referenced in the poem.\
                Explain the technical characteristics of that language.\
                Describe how those characteristics are used as metaphors.\
                Connect the technical ideas to the deeper meaning of the poem.\
                Keep the explanation clear and structured."),
    ("human", "{poem}"),
])

critic_chain = critic_prompt | chat

final_chain = {"poem": author_chain} | critic_chain

content = final_chain.invoke({
    "language": "rust"
}).content

print(content)
```
